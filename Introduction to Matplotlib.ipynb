## Matplotlib example workflow

```python
# o.import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1, 2, 3, 4]
y = [10, 20, 35, 50]

# 2. set up a plot
fig, ax = plt.subplots()

# 3. plot
ax.plot(x, y)

# 4. customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. show the plot
plt.show()
```

### two options of subplot

```python
# subplot option one
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
# plot to different axes
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),height=nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

##  whashuold we use between oo-method and pyplot
* when plotting something quickliy it's okay to use pyplot
* when plotting something more advanced use the oo method

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
heart_disease=pd.read_csv(r"C:\Users\user\Documents\[TutsNode.com] - Complete Machine Learning & Data Science Bootcamp 2021\8. Matplotlib Plotting and Data Visualization\13.1 heart-disease.csv")
heart_disease.head()
```

```python

over_50=heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# oo method mixed with pyplot
fig, ax=plt.subplots(figzize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target"
            ax=ax);

ax.set_xlim([45,50])
```

```python
# set up astyle
plt.style.use('seaborn-v0_8-whitegrid')
## oo method from scratch
over_50=heart_disease[heart_disease['age']>50]
over_50.head()
# method from scratch
fig, ax=plt.subplots(figsize=(10,6))
# plot the data
scatter=ax.scatter(x=over_50['age'],
                  y=over_50['chol'],
                  c=over_50['target']);
# Customize data
ax.set(title="HEAT DESEASE AND CHORESTOL",
      xlabel="Age",
      ylabel="Chol")
# Add a legend
ax.legend(*scatter.legend_elements(),title="target");
# Add a horizontal line
ax.axhline(over_50["chol"].mean());
```

```python
# subplots of chol,thalach and age
fig,(ax0,ax1)=plt.subplots(nrows=2,
                          ncols=1,
                          figsize=(10,6))
# ADD data to ax0
ax0.scatter(x=over_50["age"],
           y=over_50["chol"],
           c=over_50["target"]);
# customize ax0
ax0.set(title="Heart desease",
       xlabel="Age",
       ylabel="chol");
# Add data to ax1
# ADD data to ax0
ax1.scatter(x=over_50["age"],
           y=over_50["thalach"],
           c=over_50["target"]);
# customize ax1
ax0.set(title="Heart desease and thalach",
       xlabel="Age",
       ylabel="thalach");
```
